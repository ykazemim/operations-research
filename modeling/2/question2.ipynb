# Part 1

```python
import pyomo.environ as pyo

# Model
model = pyo.ConcreteModel()

# Set I
model.I = pyo.Set(initialize=[1, 2, 3, 4, 5])

# Decision Variables
model.x = pyo.Var(model.I, model.I, domain=pyo.NonNegativeReals)

# Objective function
model.obj = pyo.Objective(
    expr=91000 * model.x[2, 1] + 96000 * model.x[3, 1] + 25000 * model.x[4, 1] + 115000 * model.x[5, 1],
    sense=pyo.maximize
)

# Constraints
def constraint_sigma_x5(model, i):
    return sum(model.x[5, j] for j in model.I) == 50
model.sigma_x5 = pyo.Constraint(model.I, rule=constraint_sigma_x5)

def constraint_1(model):
    return 1.05 * model.x[3, 2] + 0.27 * model.x[4, 2] + 1.26 * model.x[5, 2] == model.x[2, 1] + model.x[2, 3] + model.x[2, 4]
model.constraint_1 = pyo.Constraint(rule=constraint_1)

def constraint_2(model):
    return 0.95 * model.x[2, 3] + 0.26 * model.x[4, 3] + 1.2 * model.x[5, 3] == model.x[3, 1] + model.x[3, 2] + model.x[3, 4]
model.constraint_2 = pyo.Constraint(rule=constraint_2)

def constraint_3(model):
    return 3.67 * model.x[2, 4] + 3.85 * model.x[3, 4] + 4.63 * model.x[5, 4] == model.x[4, 1] + model.x[4, 2] + model.x[4, 3]
model.constraint_3 = pyo.Constraint(rule=constraint_3)

def constraint_xii(model, i):
    return model.x[i, i] == 0
model.xii_constraints = pyo.Constraint(model.I, rule=constraint_xii)

def constraint_xi5(model, i):
    return model.x[i, 5] == 0
model.xi5_constraints = pyo.Constraint(model.I, rule=constraint_xi5)

def constraint_x1i(model, i):
    return model.x[1, i] == 0
model.x1i_constraints = pyo.Constraint(model.I, rule=constraint_x1i)

# Solve the model
solver = pyo.SolverFactory('glpk')
result = solver.solve(model, tee=False)

print(result)

# Display the results
model.display()

# Display results
# model.obj.display()
# for i in model.I:
#     for j in model.I:
#         print(f"x_{i},{j} = {model.x[i, j].value}")
```
